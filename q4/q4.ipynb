```python
import csv
import matplotlib.pyplot as plt
import numpy as np

if __name__ == '__main__':

    f=open('q4.csv',encoding='ANSI')

    from matplotlib import font_manager,rc
    font_path = 'NanumGothicET.ttf'
    font = font_manager.FontProperties(fname=font_path).get_name()
    rc('font', family=font)

    data=csv.reader(f)
    header=next(data)
    header=next(data)

    dic_u={}
    dic_d={}
    dic_a={}
    for row in data:
        if row[3] not in dic_a:
            dic_a[row[3]]=int(row[10].replace(',',''))+int(row[11].replace(',',''))+int(row[12].replace(',',''))+int(row[13].replace(',',''))
        else:
            dic_a[row[3]]=dic_a.get(row[3])+int(row[10].replace(',',''))+int(row[11].replace(',',''))+int(row[12].replace(',',''))+int(row[13].replace(',',''))
        
        if row[3] not in dic_u:
            dic_u[row[3]]=int(row[10].replace(',',''))+int(row[12].replace(',',''))
        else:
            dic_u[row[3]]=dic_u.get(row[3])+int(row[10].replace(',',''))+int(row[12].replace(',',''))
        
        if row[3] not in dic_d:
            dic_d[row[3]]=+int(row[11].replace(',',''))+int(row[13].replace(',',''))
        else:
            dic_d[row[3]]=dic_d.get(row[3])+int(row[11].replace(',',''))+int(row[13].replace(',',''))
    f.close()

    dic_1=sorted(dic_u.items(), key=lambda x: x[1], reverse=True)
    sub=[]
    user=[]
    colors = ['#FF0000', '#00FF00', '#0000FF', '#FFFF00', '#FF00FF', '#00FFFF', '#800000', '#008000', '#000080', '#808000',
          '#800080', '#008080', '#FFA500', '#FFC0CB', '#FFD700', '#FFFFE0', '#00FF7F', '#7FFF00', '#00FFFF', '#FF4500',
          '#8A2BE2', '#7B68EE', '#FF69B4', '#FF1493', '#BA55D3', '#B0E0E6', '#48D1CC', '#32CD32', '#FF8C00', '#F08080']


    for i in dic_1:
        sub.append(i[0])
        user.append(i[1])
    fig = plt.figure(figsize=(60, 20))
    plt.bar(sub[:30],user[:30],label=sub[:30],align='center',width=0.8,color=colors,edgecolor='black')
    plt.title("",fontsize=50)
    plt.xticks(sub[:30],rotation=45, fontsize=25)
    plt.yticks(fontsize=30)
    plt.legend(prop={'size': 25},ncol=2)
    plt.subplots_adjust(left=0.05, right=0.95)
    plt.show()

    fig = plt.figure(figsize=(6, 4))

    dic_1=sorted(dic_d.items(), key=lambda x: x[1], reverse=True)
    sub=[]
    user=[]

    for i in dic_1:
        sub.append(i[0])
        user.append(i[1])
    fig = plt.figure(figsize=(60, 20))
    plt.bar(sub[:30],user[:30],label=sub[:30],align='center',width=0.8,color=colors,edgecolor='black')
    plt.title("",fontsize=50)
    plt.xticks(sub[:30],rotation=45, fontsize=25)
    plt.yticks(fontsize=30)
    plt.legend(prop={'size': 25},ncol=2)
    plt.subplots_adjust(left=0.05, right=0.95)
    plt.show()

    fig = plt.figure(figsize=(6, 4))

    dic_1=sorted(dic_a.items(), key=lambda x: x[1], reverse=True)
    sub=[]
    user=[]

    for i in dic_1:
        sub.append(i[0])
        user.append(i[1])
    fig = plt.figure(figsize=(60, 20))
    plt.bar(sub[:30],user[:30],label=sub[:30],align='center',width=0.8,color=colors,edgecolor='black')
    plt.title("",fontsize=50)
    plt.xticks(sub[:30],rotation=45, fontsize=25)
    plt.yticks(fontsize=30)
    plt.legend(prop={'size': 25},ncol=2)
    plt.subplots_adjust(left=0.05, right=0.95)
    plt.show()
```
