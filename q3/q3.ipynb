```python
import csv
import matplotlib.pyplot as plt
import numpy as np

if __name__=="__main__":


    f=open('2008_2014.csv','r',encoding='ansi')
    data=csv.reader(f)
    header=next(data)
    rows=[]
    
    for row in data:
        rows.append(row)
    
    data_2008_2014 = []
    temp_array = []
    for i in range(1, len(rows[0])):
        temp_array.append(rows[0][i])
        if i % 5 == 0:
            data_2008_2014.append(temp_array)
            temp_array = []
    if temp_array:
        data_2008_2014.append(temp_array)
    f.close()

    f=open('2015_2022.csv','r',encoding='ansi')
    data=csv.reader(f)
    header=next(data)
    rows=[]
    
    for row in data:
        rows.append(row)
    
    data_2015_2022 = []
    temp_array = []
    for i in range(1, len(rows[0])):
        temp_array.append(rows[0][i])
        if i % 5 == 0:
            data_2015_2022.append(temp_array)
            temp_array = []
    if temp_array:
        data_2015_2022.append(temp_array)
    f.close()

    data=data_2008_2014+data_2015_2022
    year=[2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]

    male=[]
    female=[]
    for i in data:
        male.append(-int(i[2].replace(',','')))
        female.append(int(i[3].replace(',','')))
    all= -(np.array(male, dtype=int)) + np.array(female, dtype=int)

    fig, ax = plt.subplots()
    ax.set_title("graph of the population of female")
    ##ax.barh(year, male, height=-0.4, label="male",color='b')
    ax.bar(year, all, width=-0.4, align='center', label='all', color='g')
    ax.bar(year, female, width=0.4, align='center', label='female', color='r')
    ax.set_xlabel("the number of people")
    ax.set_ylabel("year")
    ax.legend()


    plt.show()


    ch=[]
    for i in data:
        ch.append(float(i[4].replace(' ','')))
    
    fig = plt.figure()

    plt.scatter(year,ch)
    plt.title("difference rate")
    plt.xlabel("year")
    plt.ylabel("rate")
    plt.show()
    
```
