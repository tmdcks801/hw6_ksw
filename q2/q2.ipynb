```python
##주사위 생성후 저장

import random
import csv

def roll_dice():
    return random.randint(1, 6)
if __name__=="__main__":
    d_100=[100,0,0,0,0,0,0]
    d_1000=[1000,0,0,0,0,0,0]
    d_10000=[10000,0,0,0,0,0,0]
    d_100000=[100000,0,0,0,0,0,0]
    for i in range(100):
        dice_roll = roll_dice()
        d_100[dice_roll]+=1
    for i in range(1000):
        dice_roll = roll_dice()
        d_1000[dice_roll]+=1
    for i in range(10000):
        dice_roll = roll_dice()
        d_10000[dice_roll]+=1
    for i in range(100000):
        dice_roll = roll_dice()
        d_100000[dice_roll]+=1

    head=["횃수",1,2,3,4,5,6]
    f=open('dice.csv','w',newline='')
    ww=csv.writer(f)
    ww.writerow(head)
    ww.writerow(d_100)
    ww.writerow(d_1000)
    ww.writerow(d_10000)
    ww.writerow(d_100000)
    f.close()

```
